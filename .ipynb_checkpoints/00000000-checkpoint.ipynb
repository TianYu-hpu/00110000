这是一个自问自答的过程，我将用0/1这两个最简单的数字告诉大家这个原理，这是我对宇宙的另一种认知维度的一个剖析。

那是不是这个速度我们能达到呢？我在这里要说一下答案，不能，或者说以后都不能，宇宙是被【造宇者】设计出来的，ta在设计的时候已经精确计算了咱们现在正在讨论的这个话题。

回到第00问题答案，宇宙为什么得有尽头，在做一个系统的时候我们得知道这个系统的边界在哪里，不是为了让我们的系统没有边界，而是让我们的系统运行在一个可以承受的最大可承受范围内。这就是做人做事要有边界的重要性。
良好稳定的运行，社会也是这样子的，显然因为新冠肺炎，我们现在的整个人类社会都不太稳定。跑题儿了！
